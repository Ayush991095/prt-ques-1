```python
from datetime import datetime
df['date'] = pd.to_datetime(df['date'])
```

```python
df.info()
```

```python
df['day_of_week'] = df['date'].dt.dayofweek

```

```python
df['date_only'] = df['date'].dt.date
```

```python
df
```

```python
daily_appliances = df.groupby('date_only')['appliances'].sum().reset_index()
```

```python
daily_appliances['day_of_week'] = pd.to_datetime(daily_appliances['date_only']).dt.dayofweek
daily_appliances['is_weekend'] = daily_appliances['day_of_week'].isin([5, 6])
```

```python
weekend_data = daily_appliances[daily_appliances['is_weekend']]
weekday_data = daily_appliances[~daily_appliances['is_weekend']]
```

```python
weekend_top = weekend_data['appliances'].quantile(0.90)
weekday_bottom = weekday_data['appliances'].quantile(0.50)
```

```python
high_weekends = weekend_data[weekend_data['appliances'] >= weekend_top]
matching_dates = high_weekends[high_weekends['appliances'] <= weekday_bottom]
```
